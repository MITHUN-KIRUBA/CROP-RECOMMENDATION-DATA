````python
# Step 8: Advanced Data Analysis
from pyspark.sql.functions import *

print("üåæ CROP DISTRIBUTION ANALYSIS:")
crop_dist = df_spark.groupBy("Suggested_Crop").agg(
    count("*").alias("Count"),
    avg("Expected_Yield").alias("Avg_Yield"),
    stddev("Expected_Yield").alias("StdDev_Yield")
).orderBy(desc("Count"))

crop_dist.show()

print("\nüèûÔ∏è STATE-WISE ANALYSIS:")
state_analysis = df_spark.groupBy("State").agg(
    count("*").alias("Records"),
    countDistinct("Suggested_Crop").alias("Unique_Crops"),
    avg("Expected_Yield").alias("Avg_Yield")
).orderBy(desc("Records"))

state_analysis.show()

print("\nüìÖ SEASONAL ANALYSIS:")
seasonal_analysis = df_spark.groupBy("Season").agg(
    count("*").alias("Records"),
    avg("Expected_Yield").alias("Avg_Yield")
).orderBy(desc("Avg_Yield"))

seasonal_analysis.show()

print("\nüå± SOIL TYPE ANALYSIS:")
soil_analysis = df_spark.groupBy("Soil_Type").agg(
    count("*").alias("Records"),
    avg("Soil_pH").alias("Avg_pH"),
    avg("Expected_Yield").alias("Avg_Yield")
).orderBy(desc("Avg_Yield"))

soil_analysis.show()
````

````python
# Step 9: Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("üìä DATA VISUALIZATIONS")

# Convert to pandas for visualization
crop_dist_pd = crop_dist.toPandas()
state_analysis_pd = state_analysis.toPandas()

# Create visualizations
plt.figure(figsize=(15, 10))

# 1. Crop Distribution
plt.subplot(2, 2, 1)
plt.bar(crop_dist_pd['Suggested_Crop'], crop_dist_pd['Count'])
plt.title('Crop Distribution')
plt.xticks(rotation=45)

# 2. Average Yield by Crop
plt.subplot(2, 2, 2)
plt.bar(crop_dist_pd['Suggested_Crop'], crop_dist_pd['Avg_Yield'])
plt.title('Average Yield by Crop')
plt.xticks(rotation=45)

# 3. State-wise Records
plt.subplot(2, 2, 3)
plt.bar(state_analysis_pd['State'], state_analysis_pd['Records'])
plt.title('Records by State')
plt.xticks(rotation=45)

# 4. Average Yield by State
plt.subplot(2, 2, 4)
plt.bar(state_analysis_pd['State'], state_analysis_pd['Avg_Yield'])
plt.title('Average Yield by State')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

````python
# Step 10: Feature Engineering for ML
print("üîß FEATURE ENGINEERING")
from pyspark.ml.feature import StringIndexer, VectorAssembler, OneHotEncoder
from pyspark.ml import Pipeline

# Convert categorical variables to numerical
soil_indexer = StringIndexer(inputCol="Soil_Type", outputCol="Soil_Type_Index")
state_indexer = StringIndexer(inputCol="State", outputCol="State_Index")
season_indexer = StringIndexer(inputCol="Season", outputCol="Season_Index")
crop_indexer = StringIndexer(inputCol="Suggested_Crop", outputCol="Crop_Index")

# One-hot encoding
soil_encoder = OneHotEncoder(inputCol="Soil_Type_Index", outputCol="Soil_Type_Vec")
state_encoder = OneHotEncoder(inputCol="State_Index", outputCol="State_Vec")
season_encoder = OneHotEncoder(inputCol="Season_Index", outputCol="Season_Vec")

# Feature columns
feature_columns = ["Soil_pH", "Nitrogen", "Phosphorus", "Potassium",
                  "Temperature", "Humidity", "Rainfall",
                  "Soil_Type_Vec", "State_Vec", "Season_Vec"]

assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")

print("‚úÖ Feature engineering pipeline defined")
````

````python
# Step 11: Prepare Data for ML Models
print("ü§ñ PREPARING DATA FOR MACHINE LEARNING")

# Create preprocessing pipeline
preprocessing_pipeline = Pipeline(stages=[
    soil_indexer, state_indexer, season_indexer,
    soil_encoder, state_encoder, season_encoder,
    assembler
])

# Fit the pipeline
pipeline_model = preprocessing_pipeline.fit(df_spark)
df_processed = pipeline_model.transform(df_spark)

print("‚úÖ Data preprocessing completed")
print("Processed data sample:")
df_processed.select("features", "Expected_Yield", "Suggested_Crop").show(5)
````

````python
# Step 12: Yield Prediction Model
print("üéØ YIELD PREDICTION MODEL (Regression)")
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator

# Split data
train_data, test_data = df_processed.randomSplit([0.8, 0.2], seed=42)

print(f"Training data: {train_data.count()} records")
print(f"Testing data: {test_data.count()} records")

# Define Random Forest Regressor
rf_regressor = RandomForestRegressor(
    featuresCol="features",
    labelCol="Expected_Yield",
    numTrees=100,
    maxDepth=10,
    seed=42
)

# Train the model
print("Training Random Forest model...")
rf_model = rf_regressor.fit(train_data)
print("‚úÖ Random Forest model trained!")

# Make predictions
predictions = rf_model.transform(test_data)

# Evaluate model
evaluator = RegressionEvaluator(
    labelCol="Expected_Yield",
    predictionCol="prediction",
    metricName="rmse"
)

rmse = evaluator.evaluate(predictions)
r2 = evaluator.evaluate(predictions, {evaluator.metricName: "r2"})

print(f"üìà Model Performance:")
print(f"RMSE: {rmse:.4f}")
print(f"R¬≤ Score: {r2:.4f}")

# Show some predictions
print("\nüîç Sample Predictions:")
predictions.select("Expected_Yield", "prediction").show(10)
````

````python
# Step 13: Crop Recommendation Model
print("\nüåø CROP RECOMMENDATION MODEL (Classification)")
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Create classification pipeline
classification_pipeline = Pipeline(stages=[
    soil_indexer, state_indexer, season_indexer, crop_indexer,
    soil_encoder, state_encoder, season_encoder,
    assembler
])

# Fit pipeline
class_pipeline_model = classification_pipeline.fit(df_spark)
df_class_processed = class_pipeline_model.transform(df_spark)

# Split data
train_class, test_class = df_class_processed.randomSplit([0.8, 0.2], seed=42)

# Random Forest Classifier
rf_classifier = RandomForestClassifier(
    featuresCol="features",
    labelCol="Crop_Index",
    numTrees=100,
    maxDepth=10,
    seed=42
)

# Train model
print("Training Crop Recommendation model...")
rf_class_model = rf_classifier.fit(train_class)
print("‚úÖ Crop Recommendation model trained!")

# Predictions
class_predictions = rf_class_model.transform(test_class)

# Evaluate
class_evaluator = MulticlassClassificationEvaluator(
    labelCol="Crop_Index",
    predictionCol="prediction",
    metricName="accuracy"
)

accuracy = class_evaluator.evaluate(class_predictions)
f1 = class_evaluator.evaluate(class_predictions, {class_evaluator.metricName: "f1"})

print(f"üéØ Classification Performance:")
print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1:.4f}")
````

````python
# Step 14: Model Evaluation and Visualization
print("üìä MODEL EVALUATION AND VISUALIZATION")

# Convert to pandas for visualization
regression_pd = predictions.select("Expected_Yield", "prediction").toPandas()
classification_pd = class_predictions.select("Crop_Index", "prediction").toPandas()

# Create evaluation plots
plt.figure(figsize=(15, 5))

# Regression plot
plt.subplot(1, 3, 1)
plt.scatter(regression_pd['Expected_Yield'], regression_pd['prediction'], alpha=0.6)
plt.plot([1, 6], [1, 6], 'r--', linewidth=2)
plt.xlabel('Actual Yield')
plt.ylabel('Predicted Yield')
plt.title(f'Yield Prediction\nR¬≤ = {r2:.3f}')

# Residual plot
plt.subplot(1, 3, 2)
residuals = regression_pd['Expected_Yield'] - regression_pd['prediction']
plt.scatter(regression_pd['prediction'], residuals, alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Yield')
plt.ylabel('Residuals')
plt.title('Residual Plot')

# Classification accuracy
plt.subplot(1, 3, 3)
metrics = ['Accuracy', 'F1-Score']
scores = [accuracy, f1]
plt.bar(metrics, scores, color=['blue', 'green'])
plt.ylim(0, 1)
plt.title('Classification Performance')

plt.tight_layout()
plt.show()
````

````python
# Step 15: Fixed Prediction Interface
print("üéÆ PREDICTION INTERFACE")

def predict_optimal_crop(state, soil_type, season, soil_ph, nitrogen, phosphorus, potassium, temperature, humidity, rainfall):
    """Predict optimal crop and expected yield for given conditions"""

    # Create input data with explicit types
    input_data = [
        ("R9999", state, soil_type, float(soil_ph), int(nitrogen), int(phosphorus), int(potassium),
         float(temperature), int(humidity), int(rainfall), season, "Unknown", 0.0, 2024)
    ]

    # Define schema explicitly
    from pyspark.sql.types import StructType, StructField, StringType, DoubleType, IntegerType

    schema = StructType([
        StructField("Region_ID", StringType(), True),
        StructField("State", StringType(), True),
        StructField("Soil_Type", StringType(), True),
        StructField("Soil_pH", DoubleType(), True),
        StructField("Nitrogen", IntegerType(), True),
        StructField("Phosphorus", IntegerType(), True),
        StructField("Potassium", IntegerType(), True),
        StructField("Temperature", DoubleType(), True),
        StructField("Humidity", IntegerType(), True),
        StructField("Rainfall", IntegerType(), True),
        StructField("Season", StringType(), True),
        StructField("Suggested_Crop", StringType(), True),
        StructField("Expected_Yield", DoubleType(), True),
        StructField("Year", IntegerType(), True)
    ])

    # Create DataFrame with explicit schema
    input_df = spark.createDataFrame(input_data, schema)

    try:
        # Predict yield
        input_processed = pipeline_model.transform(input_df)
        yield_prediction = rf_model.transform(input_processed)
        predicted_yield = yield_prediction.select("prediction").collect()[0][0]

        # Predict crop
        input_class_processed = class_pipeline_model.transform(input_df)
        crop_prediction = rf_class_model.transform(input_class_processed)
        predicted_crop_idx = crop_prediction.select("prediction").collect()[0][0]

        # Get crop name from index
        crop_labels = dict(df_class_processed.select("Crop_Index", "Suggested_Crop").distinct().collect())
        predicted_crop = crop_labels.get(predicted_crop_idx, "Unknown")

        return predicted_crop, float(predicted_yield)

    except Exception as e:
        print(f"Prediction error: {e}")
        return "Unknown", 0.0

# Test the prediction function
print("üîç TEST PREDICTION:")
test_crop, test_yield = predict_optimal_crop(
    state="Punjab",
    soil_type="Loamy",
    season="Rabi",
    soil_ph=7.2,
    nitrogen=65,
    phosphorus=45,
    potassium=120,
    temperature=28.5,
    humidity=65,
    rainfall=850
)

print(f"üåæ Recommended Crop: {test_crop}")
print(f"üìà Predicted Yield: {test_yield:.2f} tons/hectare")

# Test multiple scenarios
print("\nüß™ ADDITIONAL TEST SCENARIOS:")

# Test case 1: Different conditions
test_crop1, test_yield1 = predict_optimal_crop(
    state="Kerala",
    soil_type="Clay",
    season="Kharif",
    soil_ph=6.5,
    nitrogen=80,
    phosphorus=55,
    potassium=90,
    temperature=32.0,
    humidity=85,
    rainfall=1200
)
print(f"Scenario 1 - Crop: {test_crop1}, Yield: {test_yield1:.2f}")

# Test case 2: Another scenario
test_crop2, test_yield2 = predict_optimal_crop(
    state="Maharashtra",
    soil_type="Black",
    season="Zaid",
    soil_ph=7.8,
    nitrogen=45,
    phosphorus=35,
    potassium=110,
    temperature=35.0,
    humidity=50,
    rainfall=600
)
print(f"Scenario 2 - Crop: {test_crop2}, Yield: {test_yield2:.2f}")
````
